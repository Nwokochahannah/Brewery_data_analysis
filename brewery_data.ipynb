# UTIVA QUESTIONS SECTION A

```python
# Total profit of the brewery
total_profit = df_copy["PROFIT"].sum()
total_profit
```

```python
# Profit of the two territories
territories = []
for i in df_copy["COUNTRIES"]:
    if i == "Nigeria" or i == "Ghana":
        territories.append("Anglophone")
    else:
        territories.append("Francophone")
df_copy["TERRITORY"] = territories

territory = df_copy["TERRITORY"].unique()
anglophone = []
anglo = df_copy[df_copy["TERRITORY"] == "Anglophone"]
anglophone.append(anglo["PROFIT"].sum())
franco = df_copy[df_copy["TERRITORY"] == "Francophone"]
anglophone.append(franco["PROFIT"].sum())

new_file = pd.DataFrame(anglophone, territory, columns = ["Profit"])
profit_df = new_file.reset_index()
profit_df.columns = ["Territory", "Profit"]
profit_df
```

```python
def df_maker(df, col, val):
    new_df = df[col == val]
    return new_df
df_maker(df_copy, df_copy.COUNTRIES, "Ghana").head(10)
```

```python
# Country that generated the highest profit in 2019
years_2019 = df_maker(df_copy, df_copy.YEARS, 2019)
group_country = years_2019.groupby("COUNTRIES") ["PROFIT"].sum().reset_index()
answer_country = group_country.iloc[group_country["PROFIT"].idxmax()]
answer_country["COUNTRIES"]
```

```python
# YEAR WITH HIGHEST PROFIT
group_year = df_copy.groupby("YEARS") ["PROFIT"].sum().reset_index()
answer_year = group_year.iloc[group_year["PROFIT"].idxmax()]
answer_year["YEARS"]
```

```python
# MONTH WITH THE LEAST PROFIT
group_month = df_copy.groupby("MONTHS") ["PROFIT"].sum().reset_index()
answer_month = group_month.iloc[group_month["PROFIT"].idxmin()]
answer_month["MONTHS"]
```

```python
# MINIMUM PROFIT IN THE MONTH DECEMBER 2018
year_2018 = df_maker(df_copy, df_copy["YEARS"], 2018)
month_december = df_maker(year_2018, year_2018["MONTHS"], "December")
answer_profit = month_december["PROFIT"].min()
answer_profit
```
